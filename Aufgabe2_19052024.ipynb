Zu Beginn des Prozesses wurden die gesammelten Daten einer gründlichen Überprüfung und Bereinigung unterzogen. Ein wesentlicher Schritt bestand darin, kategoriale Variablen in binäre (Dummy-)Variablen umzuwandeln. Dieser Schritt wurde aus folgenden Gründen durchgeführt:

**2.1. Kategorienumwandlung:**

- **Erleichterung der Datenanalyse:** Die Umwandlung kategorischer Variablen in binäre Variablen erleichtert die Datenanalyse, insbesondere bei der Anwendung von statistischen Modellen und maschinellem Lernen.

- **Verbesserung der Modellgenauigkeit:** Durch die Umwandlung von kategorischen Variablen in binäre Variablen können komplexe Zusammenhänge zwischen den Variablen besser erfasst werden, was wiederum die Modellgenauigkeit verbessern kann.

**2.2. Hinzufügen neuer Variablen:**

Im Zuge der Datenbereinigung und -umwandlung wurden auch neue Variablen hinzugefügt, darunter:

- **Hauptstadt (Capital):** Diese binäre Variable gibt an, ob eine beobachtete Stadt die Hauptstadt eines Landes ist oder nicht. - **Große Stadt (Large City):** Diese binäre Variable gibt an, ob eine beobachtete Stadt als groß eingestuft wird oder nicht, basierend auf der Einwohnerzahl. Die Entscheidung, diese Variable hinzuzufügen, wurde getroffen, da die Größe einer Stadt eine Rolle bei der Bestimmung von Wohnungspreisen spielen kann. Größere Städte haben oft eine höhere Nachfrage nach Wohnraum und können daher höhere Preise aufweisen.

- **Küstenlage (Coastal Location):** Diese binäre Variable gibt an, ob eine beobachtete Stadt an der Küste liegt oder nicht. Die Entscheidung, diese Variable hinzuzufügen, wurde getroffen, da Küstenstädte oft eine einzigartige Attraktivität haben und daher Unterschiede in den Wohnungspreisen aufweisen können im Vergleich zu nicht-küstennahen Städten.

```python

#hier werden die kategoriale Variable Elevator in der Spalte in das binäre (Dummy) Variable konvertiert, the origin variable Elevator will be deleted the new boolean Variable Elevator_Not available will be added to the Data Frame
VARIABLES_TO_OHE = ["Elevator"]

df_Copy=pd.concat([df_Copy,pd.get_dummies(df_Copy[VARIABLES_TO_OHE], drop_first=True)], axis=1)

```

```python
#Delete the variable we do not need for ML
df_Copy=df_Copy.drop(['ID','Property Type', 'Elevator' ], axis=1)
```

```python
# Erstelle eine neue Variable 'Capital', um anzuzeigen, ob sich die Immobilie in Lissabon befindet, 
# da dieser Faktor die Immobilienpreise maßgeblich beeinflussen kann. 
# Diese binäre Variable wird bei der Analyse der Preistreiber helfen.

city = df_Copy['Location']

def check(row):
    return 'Lisboa' in row['Location']

df_Copy['Capital'] = df_Copy.apply(lambda row: check(row), axis=1)
df_Copy.head()
```

```python
# Erstelle neue Variablen 'Metropolitan Areas' und 'Large cities' (für Städte mit über 100.000 Einwohnern), um anzuzeigen, ob sich die Immobilie in einem Ballungsraum und/oder in einer Großstadt befindet, 
# da diese Faktoren die Immobilienpreise maßgeblich beeinflussen können. 
# Diese binären Variablen werden bei der Analyse der Preistreiber helfen.

city = df_Copy['Location']

def check(row):
    return any(city in row['Location'] for city in ['Lisboa', 'Amadora', 'Queluz', 'Setúbal', 'Almada', 'Agualva-Cacém','Porto', 'Vila Nova de Gaia', 'Gondomar', 'Rio Tinto', 'Póvoa de Varzim', 'Matosinhos'])
def checkCity(row):
    return any(city in row['Location'] for city in ['Lisboa', 'Porto', 'Braga', 'Setúbal', 'Vila Nova de Gaia', 'Amadora','Coimbra','Funchal']) 
def checkCoastal(row):
    return any(city in row['Location'] for city in ['Albufeira', 'Aljezur', 'Armação de Pêra', 'Aveiro', 'Cacela Velha', 'Carvoeiro', 'Cascais', 'Comporta', 'Costa da Caparica', 'Ericeira', 'Faro', 'Ferragudo', 'Figueira da Foz','Lagos', 'Nazaré','Óbidos', 'Odeceixe', 'Olhão', 'Peniche', 'Portimão', 'Porto', 'Porto Covo','Sagres','Sesimbra','Tavira', 'Viana do Castelo'])

  
df_Copy['Metropolitan Areas'] = df_Copy.apply(lambda row: check(row), axis=1)
df_Copy['Large cities']=df_Copy.apply(lambda row: checkCity(row), axis=1)
df_Copy ['Coastal Locations']=df_Copy.apply(lambda row: checkCoastal(row), axis=1) 
df_Copy.head()
```

```python
#excel .xlsxfile from https://www.ine.pt/xportal/xmain?xpid=INE&xpgid=ine_destaques&DESTAQUESdest_boui=594945354&DESTAQUESmodo=2 converted to csv in microsoft
#File_Age = "../data/age2022-1.csv"
File_Age= "age2022-1.csv"
df2 = pd.read_csv(File_Age,
                sep=';',
                encoding='ISO-8859-1',
                header=0, 
                on_bad_lines='skip',
                
                engine='python'
                )
df2.head()
```

```python
#Ensure the location columns are in the correct format
df_Copy['Location'] = df_Copy['Location'].astype(str)
df2['Designação'] = df2['Designação'].astype(str)

# Initialize the 'Mean Age' column in df1

df_Copy['Median Age'] = None

# Iterate over the rows of df1
for i, row in df_Copy.iterrows():
    for j, location in df2['Designação'].items():
        if location in row['Location']:
            df_Copy.at[i, 'Median Age'] = df2.at[j,'Idade mediana da população residente']
            break

# Check the updated DataFrame
print(df_Copy.head())
```

Der nächste Schritt ist die Integration von Entfernungsdaten aus der Datei "location_distances-3.csv" in den Hauptdatensatz. Die Entfernungsdaten wurden mithilfe der Google Maps API erfasst und sollen als neue Spalte "Distance to Nearest Coastal Point (km)" in den Hauptdatensatz integriert werden. Der Hauptdatensatz wird als DataFrame df_Copy bezeichnet.

```python
# Hier werden die Daten aus der CSV-Datei "location_distances-3.csv" eingelesen, die mithilfe der Google Maps API gesammelt wurden.
# Es wird angenommen, dass die Daten im CSV-Format vorliegen, wobei ';' als Trennzeichen verwendet wird und die Kodierung 'ISO-8859-1' ist.
# Die erste Zeile der Datei wird als Header verwendet, und fehlerhafte Zeilen werden übersprungen.
# Der pandas-Parser wird verwendet, um die Datei zu lesen.

File_Distance= "location_distances-3.csv"
df3 = pd.read_csv(File_Distance,
                sep=';',
                encoding='ISO-8859-1',
                header=0, 
                on_bad_lines='skip',
                
                engine='python'
                )
df3.tail()
```

```python
#Ensure the location columns are in the correct format
df_Copy['Location'] = df_Copy['Location'].astype(str)
df3['LocationMap'] = df3['Location'].astype(str)

# Initialize the 'Distance To Ocean' column in df1

df_Copy['Distance To Ocean'] = None

# Iterate over the rows 
for i, row in df_Copy.iterrows():
    for j, location in df3['LocationMap'].items():
        if location in row['Location']:
            df_Copy.at[i, 'Distance To Ocean'] = df3.at[j,'Distance to Nearest Coastal Point (km)']
            break

# Check the updated DataFrame
print(df_Copy.head())
df_Copy.info()
```
