```python
# Laden der notwendigen Bibliotheken
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns   # https://seaborn.pydata.org

# Make sure the plots are displayed inline (within the notebook)
%matplotlib inline
import numpy as np
import csv

FILENAME_DATASET = "../data/property_listingsERA5.csv"
#fs ="c:/Users/sve75036/Downloads/cleaned_property_listingsERA.csv"
```

```python
import chardet

# Detect file encoding
with open(FILENAME_DATASET, 'rb') as file:
    result = chardet.detect(file.read())
    encoding = result['encoding']
    print(f"Detected encoding: {encoding}")
```

```python
#Datensatz wird kopiert und  kontroliert ob die Dimension des Datensatzes ist korrekt
fields = ["ID", "Location", "Price", "Property Type", "Rooms", "WCs", "Net Floor Area (m2)", "Gross Private Area (m2)", 
                        "Parking", "Floor", "Elevator", "Energy Performance Certificate", ]
df = pd.read_csv(FILENAME_DATASET,
                sep=';',
                encoding='ISO-8859-1',
                header=0, 
                quotechar='"',
                quoting=csv.QUOTE_NONNUMERIC,
                on_bad_lines='skip',
                names=fields,
                )
df_Copy = df
nSpalten = df_Copy.shape[1]
nZeilen = df_Copy.shape[0]
if nSpalten == 12 and nZeilen == 1949:
    print(
       f"Der Datensatz hat  Zeilen {nZeilen} und {nSpalten} Spalten. Die Dimension des Datensatzes ist korrekt.")
else:
    print("Fehler: Der Datensatz hat nicht die erwartete Dimension", nSpalten, nZeilen)
df_Copy.head()
```

```python

#hier werden die kategoriale Variable Elevator in der Spalte in das binäre (Dummy) Variable konvertiert, the origin variable Elevator will be deleted the new boolean Variable Elevator_Not available will be added to the Data Frame
VARIABLES_TO_OHE = ["Elevator"]

df_Copy=pd.concat([df_Copy,pd.get_dummies(df_Copy[VARIABLES_TO_OHE], drop_first=True)], axis=1)

```

```python
#Zeigt eine Zusammenfassung der Informationen über den Data Frame an, einschließlich Anzahl der Nicht-Null-Einträge und Datentypen der Spalten
df_Copy.info()
```

```python
columnsNum =['Parking', 'Rooms','WCs','Net Floor Area (m2)','Gross Private Area (m2)', 'Floor']
for column in columnsNum:
    df_Copy[column]= pd.to_numeric(df_Copy[column], errors='coerce')
    df_Copy[column].abs()
print(df_Copy.head())
```

```python
#Delete the variable we do not need for ML
df_Copy=df_Copy.drop(['ID','Property Type', 'Elevator' ], axis=1)
```

```python
# Create a new variable 'Capital' to denote whether the property is located in Lisbon, 
# as this factor can significantly influence property prices. This binary variable 
# will help in the analysis of price determinants.
city = df_Copy['Location']

def check(row):
    return 'Lisboa' in row['Location']

df_Copy['Capital'] = df_Copy.apply(lambda row: check(row), axis=1)
df_Copy.head()
```

```python
# Create a new variable 'Metropolitan Areas' and 'Large cities'(for cities over 100000 population) to denote whether the property is located in metropolitan area or/and in a cit, 
# as this factors can significantly influence property prices. These binary variables 
# will help in the analysis of price determinants.
city = df_Copy['Location']

def check(row):
    return any(city in row['Location'] for city in ['Lisboa', 'Amadora', 'Queluz', 'Setúbal', 'Almada', 'Agualva-Cacém','Porto', 'Vila Nova de Gaia', 'Gondomar', 'Rio Tinto', 'Póvoa de Varzim', 'Matosinhos'])
def checkCity(row):
    return any(city in row['Location'] for city in ['Lisboa', 'Porto', 'Braga', 'Setúbal', 'Vila Nova de Gaia', 'Amadora','Coimbra','Funchal']) 

df_Copy['Metropolitan Areas'] = df_Copy.apply(lambda row: check(row), axis=1)
df_Copy['Large cities']=df_Copy.apply(lambda row: checkCity(row), axis=1)
df_Copy.head()
```
